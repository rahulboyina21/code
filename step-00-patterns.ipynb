## How can you manipulate Rows / Columns using a expression or a equation to give us the expected result in a sequence.




all the rows and columns are zero with the x rows and x colums

```python
for i in range(5):
    for j in range(5):
        print("0",end="")
    print()
```

Pattern 2 manipulation of i=j like a triangle 



```python
for i in range(5):
    for j in range(i):
        print("0",end="")
    print()

for i in range(5):
    for j in range(i):
        print(j,end="")
    print()
```

Pattern 03:


```python
for i in range(5):
    for j in range(i+1):
        print(j+1,end="")
    print()
```

Pattern 4:



```python
for i in range(6):
    for j in range(i+1):
        print(i+1,end="")
    print()
```

Pattern 5:




	•	chr() → number to character
	•	ord() → character to number

```python
for i in range(65,69):
    for j in range(65,i+1):
        print(chr(i),end="")
    print()
```

Pattern 6:





```python
for i in range(5):
    for j in range(i+1):
        print(chr(65+i+j),end="")
    print()
```

Pattern 07:





```python
for i in range(70,64,-1):
    for j in range(i,70,1):
        print(chr(j),end="")
    print()
```
